# Gráfico da Epidemia do Ebola em Modelo SIRD
## País: Guiné
## Período: 2014-2016

```python
import pandas as pd

df= pd.read_csv('dataset/ebola/guinea_SIRD.csv')
print('Número de linhas do csv:', len(df))
dados = df.to_numpy().tolist()
print('Dados brutos do csv:', dados)

x = [12600000,0,0,0]
historico = [x]

for linha in dados:
    sird_dia = [linha[1], linha[2], linha[3], linha[4]]
    historico.append(sird_dia)

print('Dados refatorados para o modelo SIRD:', historico)

historico_normalizado=[]
for linha in historico:
    for i in range(4):
        linha[i]/=12600000
    historico_normalizado.append(linha)

print('Dados normalizados:', historico_normalizado)
```
